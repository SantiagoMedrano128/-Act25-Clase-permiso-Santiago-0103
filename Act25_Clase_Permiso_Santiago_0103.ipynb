# Clase permiso
Entender la Programación Orientada a Objetos (POO) en Python es como usar un molde para galletas: la clase es el molde y los objetos son las galletas que creas con él.

Aquí esta el código siguiendo exactamente tus requerimientos.
![link text](https://raw.githubusercontent.com/SantiagoMedrano128/-Act25-Clase-permiso-Santiago-0103/refs/heads/main/infografiaclasepermiso.png)

```python
class Permiso:
    def __init__(self):
        # Inicializamos los atributos vacíos
        self.nombre = ""
        self.nivel = 0
        self.descripcion = ""

    def leerdatos(self):
        """Solicita al usuario los datos para el permiso."""
        print(f"\n--- Ingresando datos del permiso ---")
        self.nombre = input("Nombre del permiso: ")
        self.nivel = input("Nivel de acceso (numérico): ")
        self.descripcion = input("Descripción breve: ")

    def mostrardatos(self):
        """Muestra la información actual del objeto."""
        print(f"\n >>> FICHA DE PERMISO <<<")
        print(f"Nombre: {self.nombre}")
        print(f"Nivel: {self.nivel}")
        print(f"Descripción: {self.descripcion}")
        print("-" * 25)

    def validar_permiso(self):
        """Tercera función: Simula una validación del sistema."""
        print(f"SISTEMA: Validando el permiso '{self.nombre}'...")
        print(f"RESULTADO: El acceso de nivel {self.nivel} ha sido activado correctamente.")

# --- Creación de Objetos (Instancias) ---

# Objeto 1: Interactivo (Tú ingresas los datos)
print("Configuración del Permiso 1")
permiso1 = Permiso()
permiso1.leerdatos()

# Objeto 2: Datos de ejemplo predefinidos
print("\nConfiguración del Permiso 2 (Automática)")
permiso2 = Permiso()
permiso2.nombre = "Admin_Sistema"
permiso2.nivel = "10"
permiso2.descripcion = "Control total sobre los archivos del servidor"

# --- Ejecución de Funciones ---

print("\n" + "="*30)
print("REPORTE FINAL DE OBJETOS")
print("="*30)

# Acciones del Permiso 1
permiso1.mostrardatos()
permiso1.validar_permiso()

# Acciones del Permiso 2
permiso2.mostrardatos()
permiso2.validar_permiso()

print(">> ¡Gracias por su visita!")
print("Programa realizado por Santiago Medrano NL = 37")
```
